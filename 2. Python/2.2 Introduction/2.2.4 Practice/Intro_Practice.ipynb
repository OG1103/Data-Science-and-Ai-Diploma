## 5. List Manipulation

````python
list_one = [1,2,3,4,"amit","learning",True,[1,2,3,"Ture",["m1","d1","ds"]]]

print(list_one)
print(type(list_one))
print(len(list_one))

print(list_one[7][1]) # Print data in that index

list_one[7][1] = 4 # Change Data in that index

print(list_one[7][1]) # Print data in that index to test change

print(list("amit")) # Use List Function to convert any datatype to list 

list_one.append("Machine Learning") # Add Machine Learning at the end 
print(list_one)
list_one.insert(1,"omar Gaber") # Add Item in position 1 (indexed insert)
print(list_one)

list_one.extend(["1.list_2","2. Testing"]) # Extend an array ontop of list_one. So list_one is now the combo between List_one and the extended list 
print(list_one)

list_one.remove("2. Testing") # Remove this specific element
print(list_one)

x = list_one.pop() # pop last element from list 

print(x) # Print popped element
print(list_one) # print final_list
````

## 6. Input Function + List Manipulation

````python
fruits = ["apple","banana","cherry","kiwi"] # Create an array 

selected_fruit = input("Please Eneter Fruit you want to remove") # Ask fro which fruit to delete

if (fruits.count(selected_fruit)): # check if the element exists before we delete it 
    fruits.remove(selected_fruit) # remove element from array
else: 
    print("Not in our list")
print(fruits)
````
