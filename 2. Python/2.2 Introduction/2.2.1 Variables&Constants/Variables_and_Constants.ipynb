# 1. Variables
- A variable is a container that stores data

### Declaration (Creating a Variable)
Format: `variable_name = value`

```python
name = "Alice"
age = 25
price = 19.99
```

### Using (Calling) a Variable
- Use `print()` to display the value
- Or just write the variable name (acts as print in Jupyter)

```python
name = "Bob"

print(name)    # Using print()
name           # Just the variable name
```

### Assigning New Values
- You can change the value anytime

```python
score = 10
print(score)

score = 20      # Assign new value
print(score)

score = score + 5
print(score)
```

# 2. Constants
- A value that should NOT change
- Written in ALL CAPITAL LETTERS

### Declaration (Creating a Constant)
Format: `CONSTANT_NAME = value`

```python
PI = 3.14159
MAX_SPEED = 120
TAX_RATE = 0.15
```

### Using (Calling) a Constant
- Same as variables

```python
PI = 3.14159

print(PI)      # Using print()
PI             # Just the constant name
```

### Using Constants in Calculations

```python
PI = 3.14159
radius = 5

area = PI * radius ** 2
print("Area:", area)
```

# 3. Comparison
- **Variables**: lowercase names, can change → `age = 25`
- **Constants**: ALL_CAPS names, should not change → `PI = 3.14`
- **Calling**: Use `print()` or just the name in Jupyter
