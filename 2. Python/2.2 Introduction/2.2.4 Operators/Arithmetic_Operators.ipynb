# 1. Addition (+)
- Adds two numbers together
- Can also concatenate strings
- Can combine lists

```python
# Addition with numbers
print(10 + 5)  # 15
print(3.5 + 2.5)  # 6.0
print(-10 + 5)  # -5

# Addition with strings (concatenation)
print("Hello" + " " + "World")  # Hello World

# Addition with lists (concatenation)
print([1, 2] + [3, 4])  # [1, 2, 3, 4]
```

# 2. Subtraction (-)
- Subtracts one number from another
- Returns the difference between two numbers

```python
print(10 - 5)  # 5
print(3.5 - 2.5)  # 1.0
print(5 - 10)  # -5
print(100 - 25)  # 75
```

# 3. Multiplication (*)
- Multiplies two numbers
- Can repeat strings
- Can repeat lists

```python
# Multiplication with numbers
print(10 * 5)  # 50
print(3.5 * 2)  # 7.0
print(-5 * 3)  # -15

# Multiplication with strings (repetition)
print("Ha" * 3)  # HaHaHa
print("-" * 10)  # ----------

# Multiplication with lists (repetition)
print([1, 2] * 3)  # [1, 2, 1, 2, 1, 2]
```

# 4. Division (/)
- Divides one number by another
- Always returns a float (decimal number)
- Cannot divide by zero (raises error)

```python
print(10 / 2)  # 5.0 (always returns float)
print(15 / 4)  # 3.75
print(7 / 2)  # 3.5
print(9 / 3)  # 3.0

# Division by zero causes error
# print(10 / 0)  # ZeroDivisionError
```

# 5. Floor Division (//)
- Divides and rounds DOWN to nearest whole number
- Removes the decimal part
- Returns integer if both operands are integers

```python
print(10 // 3)  # 3 (10 ÷ 3 = 3.33, rounded down to 3)
print(15 // 4)  # 3 (15 ÷ 4 = 3.75, rounded down to 3)
print(20 // 6)  # 3
print(9 // 2)  # 4

# Comparison: regular division vs floor division
print(17 / 5)  # 3.4 (regular division)
print(17 // 5)  # 3 (floor division)
```

# 6. Modulus (%)
- Returns the remainder after division
- Useful for checking if a number is even or odd
- Useful for finding if a number is divisible by another

```python
print(10 % 3)  # 1 (10 ÷ 3 = 3 remainder 1)
print(15 % 4)  # 3 (15 ÷ 4 = 3 remainder 3)
print(20 % 5)  # 0 (20 ÷ 5 = 4 remainder 0, so 20 is divisible by 5)
print(17 % 6)  # 5

print("\n" + "="*50 + "\n")

# Check if number is even or odd
print(10 % 2)  # 0 (even number)
print(11 % 2)  # 1 (odd number)

# Check if divisible
print(15 % 3)  # 0 (15 is divisible by 3)
print(16 % 3)  # 1 (16 is NOT divisible by 3)
```

# 7. Exponentiation (**)
- Raises a number to the power of another
- Left number is the base, right number is the exponent
- Example: 2**3 means 2 × 2 × 2

```python
print(2 ** 3)  # 8 (2 × 2 × 2)
print(5 ** 2)  # 25 (5 × 5)
print(10 ** 3)  # 1000 (10 × 10 × 10)
print(3 ** 4)  # 81 (3 × 3 × 3 × 3)

# Square root (using fractional exponent)
print(16 ** 0.5)  # 4.0 (square root of 16)
print(27 ** (1/3))  # 3.0 (cube root of 27)
```

# 8. Order of Operations (PEMDAS)
- **P**arentheses ()
- **E**xponents **
- **M**ultiplication * and **D**ivision /
- **A**ddition + and **S**ubtraction -
- Operations of same priority are evaluated left to right

```python
# Without parentheses
print(2 + 3 * 4)  # 14 (multiplication first: 3*4=12, then 2+12=14)

# With parentheses
print((2 + 3) * 4)  # 20 (parentheses first: 2+3=5, then 5*4=20)

# Complex expression
print(10 + 5 * 2 ** 2)  # 30 (2**2=4, then 5*4=20, then 10+20=30)
print((10 + 5) * 2 ** 2)  # 60 (10+5=15, 2**2=4, then 15*4=60)

# Multiple operations
print(100 / 10 * 2)  # 20.0 (left to right: 100/10=10, then 10*2=20)
print(100 / (10 * 2))  # 5.0 (parentheses first: 10*2=20, then 100/20=5)
```

# 9. Assignment Operators
- Assign values to variables
- Combine arithmetic with assignment for shortcuts
- Modify variable in place

```python
# Basic assignment
x = 10
print(x)  # 10

print("\n" + "="*50 + "\n")

# Add and assign (+=)
x = 10
x += 5  # Same as: x = x + 5
print(x)  # 15

# Subtract and assign (-=)
x = 10
x -= 3  # Same as: x = x - 3
print(x)  # 7

# Multiply and assign (*=)
x = 10
x *= 2  # Same as: x = x * 2
print(x)  # 20

# Divide and assign (/=)
x = 10
x /= 2  # Same as: x = x / 2
print(x)  # 5.0

# Floor divide and assign (//=)
x = 10
x //= 3  # Same as: x = x // 3
print(x)  # 3

# Modulus and assign (%=)
x = 10
x %= 3  # Same as: x = x % 3
print(x)  # 1

# Exponent and assign (**=)
x = 2
x **= 3  # Same as: x = x ** 3
print(x)  # 8
```

# 10. Practical Examples
- Real-world use cases of arithmetic operators
- Combining multiple operators
- Common calculations

```python
# Calculate area of rectangle
length = 10
width = 5
area = length * width
print("Area:", area)  # 50

# Calculate average
num1, num2, num3 = 80, 90, 85
average = (num1 + num2 + num3) / 3
print("Average:", average)  # 85.0

# Convert Celsius to Fahrenheit
celsius = 25
fahrenheit = (celsius * 9/5) + 32
print("Temperature:", fahrenheit, "°F")  # 77.0 °F

# Calculate total price with tax
price = 100
tax_rate = 0.15  # 15%
total = price + (price * tax_rate)
print("Total with tax:", total)  # 115.0

# Calculate discount
original_price = 200
discount = 0.20  # 20% off
final_price = original_price - (original_price * discount)
print("Final price:", final_price)  # 160.0

# Split bill among friends
total_bill = 150
num_people = 4
per_person = total_bill / num_people
print("Each person pays:", per_person)  # 37.5
```
