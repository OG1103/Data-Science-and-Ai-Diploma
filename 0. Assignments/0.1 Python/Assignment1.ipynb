## Task 1 : 

- Input Validation: Check if the input string contains exactly one "@" symbol and at least one "." after the "@" symbol. If it's not a valid email, return "Invalid email".
- Extract Username: Extract and return the part of the email before the "@" symbol.
- Extract Domain: Extract and return the domain (the part between "@" and the last ".").
- Check for Domain Ending: Check if the email ends with ".com". If it does, return "Commercial Domain". If it ends with ".edu", return "Educational Domain". Otherwise, return "Other Domain".


```python
email = input("Please enter valid email")

if(email.count("@") != 1 or not email.count(".")):
    print("invalid email")
else:
    at_index = email.find("@")
    username = email[0:at_index]
    domain = email[at_index + 1:]
    if(domain.endswith(".com")):
        print("Commercial Domain")
    elif(domain.endswith(".edu")):
        print("Educational Domain")
    else:
        print("Other Domain")

    
```

## Task 2
- Encoded Message: `###!!@mocleW EPGTQ!!!6789`
- Steps to Decode:
    - Extract the core part of the message: "mocleW EPGTQ".
    - Reverse the first word: "mocleW" becomes "Welcome".
    - Replace shifted vowels in the second word:
        - "EPGTQ": No vowels to change.
    - Final decoded message: "Welcome PGTQ".

```python
encoded_message = "###!!@mocleW EPGTQ!!!6789"

core = encoded_message[6:18]

core_array = core.split(" ")
core_array[0] = core_array[0][::-1]

core_array[0] += core_array[1][0].lower()

core_array[1] = core_array[1][1:]

decoded = " ".join(core_array)

print(decoded)
```

## Task 3 

- Encoded Message: `&&&**$gnirtS PLIO!!@1234`
- Steps to Decode:
    - Extract the core part of the message: "gnirtS PLIO".
    - Reverse the first word: "gnirtS" becomes "String".
    - Replace shifted vowels in the second word:
        - "PLIO": Replace I->E and O->U to get "PLEU".
    - Final decoded message: "String PLEU".

```python
encoded_message = "&&&**$gnirtS PLIO!!@1234"

core = encoded_message[6:17]

first, sep, second = core.partition(" ")

first = first[::-1]
second = second.replace("I","E")
second = second.replace("O", "U")

decoded =  first + " " + second

print(decoded)

```

## Task 4
- Encoded Message: `##$$$@!yalpstcejorp EPUVT****9887`
- Steps to Decode:
    - Extract the core part of the message: "yalpstcejorp EPUVT".
    - Reverse the first word: "yalpstcejorp" becomes "projectplay".
    - Replace shifted vowels in the second word:
        - "EPUVT": Replace E->A, U->O to get "APOVT".
    - Final decoded message: "projectplay APTOV".

```python
encoded_message = "##$$$@!yalpstcejorp EPUVT****9887"

first_index = encoded_message.index("!")
second_index = encoded_message.index("*")

core = encoded_message[first_index + 1 : second_index]

first, sep, second = core.partition(" ")

first = first[::-1]

second = second.replace("E","A")

second = second.replace("U","O")

decoded = first + " "  + second

print(decoded)
```
