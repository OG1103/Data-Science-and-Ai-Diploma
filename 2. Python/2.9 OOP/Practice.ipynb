```python
class Person: 
    # Constructor
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def getInfo(self):
        print(f"My name is {self.name} my age is {self.age}")
    

    pass

class Engineer(Person):
    def __init__(self, name, age, major):
        super().__init__(name, age)
        self.major = major
    
    def getInfo(self):
        super().getInfo()
        print(f"Ana Mohandes")

    def printMajor(self):
        print(f"My Major is {self.major}")
        

p1 = Person("Omar",22)
e1 = Engineer("Omar",22,"met")

p1.getInfo()
e1.getInfo()
e1.printMajor()
```
