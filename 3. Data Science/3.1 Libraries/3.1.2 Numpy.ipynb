# 1. What is NumPy

NumPy (Numerical Python) is a Python library for working with numbers and mathematical operations. It provides support for large arrays and matrices, along with fast mathematical functions.

Usually more efficient in the calculations in comparison to Pandas

## Main Uses
- Performing mathematical and statistical operations
- Working with large arrays and multi-dimensional data
- Linear algebra calculations (matrices, vectors)
- Scientific computing and data processing
- Foundation for other libraries (Pandas, Scikit-learn, TensorFlow)

# 2. Importing the library 
- `import numpy as np`
- Storing the library under the shortcut `np`
